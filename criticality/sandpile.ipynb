```python
# n = 100
# grid = np.zeros((n, n))

def add_random_grain(grid, n):
    x, y = np.random.randint(0, n, 2)
    grid[x, y] += 1

def overflow(grid, n):
    for x in range(n):
        for y in range(n):
            if grid[x, y] > 3:
                grid[x, y] -= 4
                grid[(x + 1) % n, y] += 1
                grid[(x - 1) % n, y] += 1
                grid[x, (y + 1) % n] += 1
                grid[x, (y - 1) % n] += 1
```

```python
def sandpile_model(n, n_iter):
    grid = np.zeros((n, n))
    for _ in range(n_iter):
        add_random_grain(grid, n)
        overflow(grid, n)
    return grid
```

```python
grid = sandpile_model(50, 2000)
# Visualize
plt.imshow(grid, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Presence of grains')
plt.title("Sandpile Model")
plt.show()
```
